**For the given suitable table, identify the type of each attribute (nominal, ordinal, interval scaled, ratio scaled), giving justification for your choice.**

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('DataPreprocessing2.csv')
df
```

- **Roll_No - Ordinal**
- **Gender - Nominal**
- **Age - Interval**
- **Height - Ratio scaled**
- **Home_Town - Nominal**
- **Semester - Ordinal**
- **Course - Nominal**
- **Phone No - Nominal**

**1. How will you handle the missing values in the Age and Phone No attribute?**

```python
df['Age'].replace(np.nan, round(df['Age'].mean()), inplace=True)
df
```

```python
df = df[df['Phone No'].notnull()]
df
```

**2. Scale the values for all numeric attributes between 0 and 1.**

```python
df.loc[:,'Age'] = (df['Age']-df['Age'].mean())/df['Age'].std()
df.loc[:,'Height (in cms)'] = (df['Height (in cms)']-df['Height (in cms)'].mean())/df['Height (in cms)'].std()
```

```python
df
```

**3. Convert the categorical attribute Course, Gender, Home_Town and Semester in binary attributes.**

```python
# Binarizing Course column
df.loc[df['Course']=='BA(Pass)','Course'] = 0
df.loc[df['Course']=='BCom(Pass)','Course'] = 1

# Binarizing Gender column
df.loc[df['Gender']=='M','Gender'] = 0
df.loc[df['Gender']=='F','Gender'] = 1

# Binarizing Home_Town column
df.loc[df['Home_Town']=='Delhi','Home_Town'] = '000'
df.loc[df['Home_Town']=='Mumbai','Home_Town'] = '001'
df.loc[df['Home_Town']=='Shimla','Home_Town'] = '010'
df.loc[df['Home_Town']=='Agra','Home_Town'] = '011'
df.loc[df['Home_Town']=='Rajasthan','Home_Town'] = '100'

# Binarizing Semester column
df.loc[df['Semester']=='I','Semester'] = '000'
df.loc[df['Semester']=='II','Semester'] = '001'
df.loc[df['Semester']=='IV','Semester'] = '010'
df.loc[df['Semester']=='VI','Semester'] = '100'
```

#### Create various types of plots/charts like histograms, plot based on sine/cosine function based on data from a matrix. Further label different axes and data in the plot.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = np.arange(0, 10, 0.1)
```

```python
print(X)
```

```python
y = np.sin(X)
```

```python
print(y)
```

```python
plt.plot(X, y)
plt.xlabel("X")
plt.ylabel("y")
plt.title("y = sin(X)")
plt.show()
```

```python
y2 = np.cos(X)
```

```python
plt.plot(X, y2)
plt.xlabel("X")
plt.ylabel("y")
plt.title("y = cos(X)")
plt.show()
```

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.bar(names, values)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.axis('equal')
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23, 17, 35, 29, 12]
ax.pie(students, labels = langs, autopct = '%1.2f%%')
plt.show()
```

```python
a = np.array([22, 87, 5, 43, 56, 73, 55, 54, 11, 20, 51, 5, 79, 31, 27])
plt.hist(a, bins = [0, 25, 50, 75, 100])
plt.title("Histogram of result")
plt.xticks([0, 25, 50, 75, 100])
plt.xlabel('Marks')
plt.ylabel('No. of students')
plt.show()
```

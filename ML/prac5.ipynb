## Use command to compute the size of a matrix, size/length of a particular row/column, load data from a text file, store matrix data to a text file, finding out variables and their features in the current scope.

```python
import numpy as np
import pandas as pd
```

```python
A = np.array(
    [[1, 2, 3],
     [1,3,2]])
```

```python
# size of A
A.shape
```

```python
# number of elements in row 0
A[0, :].shape
```

```python
# number of elements in column 2
A[:, 2].shape
```

```python
dataset = pd.DataFrame(
    {'Column1': A[:, 0],
     'Column2': A[:, 1],
     'Column3': A[:, 2]})
dataset
```

```python
dataset.to_csv('mydata.txt', header = True, index = False, sep = ',', mode = 'a')
```

```python
data1 = pd.read_csv('mydata.txt')
data1
```

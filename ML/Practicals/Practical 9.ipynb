```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
size = 1000
data = {'X': np.random.choice(a = [100,200,300],
                              size = size,
                              p = [0.7,0.2,0.2]),
        'Y': np.random.randn(size)}

df = pd.DataFrame(data)
df
```

```python
fig, ax = plt.subplots(2,2, figsize = (15,7))

ax[0,0].hist(np.random.randn(100), bins=20, color='k', alpha=0.3)
ax[0,1].scatter(np.arange(30), np.random.randn(30))
ax[1,0].plot(np.random.randn(50), 'g--')

sns.barplot(data = df, 
            x = 'X',
            y = 'Y',
            axes = ax[1,1])

plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, StratifiedKFold
```

```python
df = pd.read_csv("IMDb database.csv")
df
```

```python
df = df.dropna()
df = df.reset_index(drop=True)
```

```python
df1 = df.copy()
df1["color_bool"] = df1["Color"].map({'Color':True ,'Black and White':False})
```

```python
df1.color_bool = df1.color_bool.fillna(False)
```

```python
df1["color_bool"].isnull().sum()
```

```python
X = df1.drop(["color_bool"], axis=1)
Y = df1.color_bool
X_Train, X_Test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state=42, shuffle =True,stratify= Y)
```
